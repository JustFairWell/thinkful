This is two collections of data by the United Nations. First is a collection of marriage statistics from different sources and runs fairly consistently from 1970 to 2010. While marriage trends have changed around the world, the trend has some more extreme examples along with some countries that have resisted and even reversed the more common trend. Second is a collection of gross national income (GNI) data and is given in a per capita form.

The goal of this report is threefold. 

1. What is the gross trend in marriage around the world? 
2. What impact, if any, does GNI have on these trends?
3. Are there any other economic factors that seem to highlight these trends?



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

marriage_df = pd.read_excel('/Users/justinsmith/Downloads/Marital status of men and women.xls', header=3)
un_gni = pd.read_csv('/Users/justinsmith/Documents/GitHub/thinkful/un_gni.csv')
```

```python
marriage_df_15 = marriage_df[['Country', 'Year', 'Marital status', 'Sex', '15-19']]
marriage_df_15 = marriage_df_15[marriage_df_15['Marital status'] == 'Single']

marriage_df_25 = marriage_df[['Country', 'Year', 'Marital status', 'Sex', '25-29']]
marriage_df_25 = marriage_df_25[marriage_df_25['Marital status'] == 'Single']

marriage_df_65 = marriage_df[['Country', 'Year', 'Marital status', 'Sex', '65+']]
marriage_df_65 = marriage_df_65[marriage_df_65['Marital status'] == 'Single']
```

Here, I have loaded my libraries, datasets and broken them down into easily referenced variables. In this case, I have extracted the age variable columns which is broken into 5 year segments (i.e. 15-19) up to 65+ years of age. I have decided to look at the perecentage of single people since this, by default, rules out
other categories like widowed and divorced, though some more recent data also has consensual unions. This gives us a better idea of who has never been married.

Now, let's take a look at some box plots, broken up by decade, to get some clarity as to what the trend of gross global trend is for marriage.

```python
marriage_70s_15 = marriage_df_15[(marriage_df_15['Year'] >= 1970) & (marriage_df_15['Year']<1980)]
marriage_80s_15 = marriage_df_15[(marriage_df_15['Year'] >= 1980) & (marriage_df_15['Year']<1990)]
marriage_90s_15 = marriage_df_15[(marriage_df_15['Year'] >= 1990) & (marriage_df_15['Year']<2000)]
marriage_00s_15 = marriage_df_15[(marriage_df_15['Year'] >= 2000) & (marriage_df_15['Year']<2010)]

fig,ax = plt.subplots(1,4, sharex=True, sharey=True)
marriage_70s_15.boxplot(column='15-19',ax=ax[0])
ax[0].set_title('1970s')
marriage_80s_15.boxplot(column='15-19',ax=ax[1])
ax[1].set_title('1980s')
marriage_90s_15.boxplot(column='15-19',ax=ax[2])
ax[2].set_title('1990s')
marriage_00s_15.boxplot(column='15-19',ax=ax[3])
ax[3].set_title('2000s')
plt.show()

marriage_70s_25 = marriage_df_25[(marriage_df_25['Year'] >= 1970) & (marriage_df_25['Year']<1980)]
marriage_80s_25 = marriage_df_25[(marriage_df_25['Year'] >= 1980) & (marriage_df_25['Year']<1990)]
marriage_90s_25 = marriage_df_25[(marriage_df_25['Year'] >= 1990) & (marriage_df_25['Year']<2000)]
marriage_00s_25 = marriage_df_25[(marriage_df_25['Year'] >= 2000) & (marriage_df_25['Year']<2010)]

fig,ax = plt.subplots(1,4, sharex=True, sharey=True)
marriage_70s_25.boxplot(column='25-29',ax=ax[0])
ax[0].set_title('1970s')
marriage_80s_25.boxplot(column='25-29',ax=ax[1])
ax[1].set_title('1980s')
marriage_90s_25.boxplot(column='25-29',ax=ax[2])
ax[2].set_title('1990s')
marriage_00s_25.boxplot(column='25-29',ax=ax[3])
ax[3].set_title('2000s')
plt.show()

marriage_70s_65 = marriage_df_65[(marriage_df_65['Year'] >= 1970) & (marriage_df_65['Year']<1980)]
marriage_80s_65 = marriage_df_65[(marriage_df_65['Year'] >= 1980) & (marriage_df_65['Year']<1990)]
marriage_90s_65 = marriage_df_65[(marriage_df_65['Year'] >= 1990) & (marriage_df_65['Year']<2000)]
marriage_00s_65 = marriage_df_65[(marriage_df_65['Year'] >= 2000) & (marriage_df_65['Year']<2010)]

fig,ax = plt.subplots(1,4, sharex=True, sharey=True)
marriage_70s_65.boxplot(column='65+',ax=ax[0])
ax[0].set_title('1970s')
marriage_80s_65.boxplot(column='65+',ax=ax[1])
ax[1].set_title('1980s')
marriage_90s_65.boxplot(column='65+',ax=ax[2])
ax[2].set_title('1990s')
marriage_00s_65.boxplot(column='65+',ax=ax[3])
ax[3].set_title('2000s')
plt.show()
```

In the 15-19 age range the most noticable trend is that the level of varience has fallen while the mean has remained relatively steady. 

In the 25-29 age range is where we see the most obvious trend. The mean percentage of single people has increased while the varience has also increased. This is also the most noteworthy range that we are currently looking at since it is closest to the modern traditional age of marriage.

In the 65+ age range the mean and variance have actually declined. 

Here is the more detailed view in summary statistics.


```python
print(marriage_70s_15.describe())
print(marriage_80s_15.describe())
print(marriage_90s_15.describe())
print(marriage_00s_15.describe())
```

Here we can see the standard deviation decreases as the mean slowly creeps up. This means that on average more people are remaining single since the mean is closer to 100% single and more valus are closer too that. 

```python
print(marriage_70s_25.describe())
print(marriage_80s_25.describe())
print(marriage_90s_25.describe())
print(marriage_00s_25.describe())
```

Here, we can see the mean marriage percentage for the age range 25-29 increases 60% relative to it's 70's mean (approx. 25% to 40%) as well as an increase of the standard deviation of 30% relative to 70's (16 points to 21 points). 

```python
print(marriage_70s_65.describe())
print(marriage_80s_65.describe())
print(marriage_90s_65.describe())
print(marriage_00s_65.describe())
```

It looks like marriage as a whole has not actually declined much over a lifetime, and the chances of being married has actually gone up in the 00's. This would suggest that people are getting married later in life.

Now let's take a look at how this compares to economic growth over the last 4 decades.

```python
un_gni_70 = un_gni[(un_gni['Year']>=1970) & (un_gni['Year']<1980)]
un_gni_80 = un_gni[(un_gni['Year']>=1980) & (un_gni['Year']<1990)]
un_gni_90 = un_gni[(un_gni['Year']>=1990) & (un_gni['Year']<2000)]
un_gni_00 = un_gni[(un_gni['Year']>=2000) & (un_gni['Year']<2010)]


fig,ax = plt.subplots(1,4, sharex=True, sharey=True)
un_gni_70.boxplot(column='Value',ax=ax[0])
ax[0].set_title('1970s')
un_gni_80.boxplot(column='Value',ax=ax[1])
ax[1].set_title('1980s')
un_gni_90.boxplot(column='Value',ax=ax[2])
ax[2].set_title('1990s')
un_gni_00.boxplot(column='Value',ax=ax[3])
ax[3].set_title('2000s')
plt.show()

print(un_gni_70.describe())
print(un_gni_80.describe())
print(un_gni_90.describe())
print(un_gni_00.describe())
```

These box plots clearly show an upward economic trend that is also verified by the numbers. The mean, in current US dollars per capita, has steadily increased from the 70's along with a greater degree of standard deviation. This means that while overall GNI has increased, some areas have reaped the rewards more than others. This can be reinforced by noticing that the mean is always higher and it's rate of growth faster than the 50 percentile. From this we can assume the distribution resembles a pareto distribution where most countries occupy a small area and smaller group of countries covers a more vast area. 

So here we have two trends that both seem to be growing. There appears, at least from a broad perspective, a chance for correlation between people getting married later and economic growth. In effort to see if we can make any stronger correlations, let's now look at some of the countries that have had some of the largest changes in marriage trends.

```python
age_list = ['20-24', '25-29', '30-34','35-39']
largest_trend = ['Sweden', 'Maldives', 'Hungary', 'Kuwait', 'United Arab Emirates', 'Belgium', 'Brazil', 'Czech Republic', 'Bahamas']

def plot_country_single(country_list, gender, marital_status):
    fig = plt.figure(figsize=(20,10))
    ax0 = fig.add_subplot(221, xlabel='Year', ylabel= 'Percentage of Single Population')
    ax1 = fig.add_subplot(222, xlabel='Year', ylabel= 'Percentage of Single Population')
    ax2 = fig.add_subplot(223, xlabel='Year', ylabel= 'Percentage of Single Population')
    ax3 = fig.add_subplot(224, xlabel='Year', ylabel= 'Percentage of Single Population')
    marriage_df_plot = marriage_df[['Country', 'Year', 'Marital status', 'Sex', '15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49','50-54', '55-59', '60-64', '65+']]
    marriage_df_plot = marriage_df_plot[marriage_df_plot['Marital status'] == marital_status]
    marriage_df_plot = marriage_df_plot[marriage_df_plot['Sex'] == gender]

    for country in country_list:
        ax0.plot(marriage_df_plot[marriage_df_plot['Country'] == country]['Year'], marriage_df_plot[marriage_df_plot['Country'] == country]['20-24'])
        ax0.set_title('Age 20-24', fontsize=20)
        ax1.plot(marriage_df_plot[marriage_df_plot['Country'] == country]['Year'], marriage_df_plot[marriage_df_plot['Country'] == country]['25-29'])
        ax1.set_title('Age 25-30', fontsize=20)
        ax2.plot(marriage_df_plot[marriage_df_plot['Country'] == country]['Year'], marriage_df_plot[marriage_df_plot['Country'] == country]['30-34'])
        ax2.set_title('Age 30-34', fontsize=20)
        ax3.plot(marriage_df_plot[marriage_df_plot['Country'] == country]['Year'], marriage_df_plot[marriage_df_plot['Country'] == country]['35-39'])
        ax3.set_title('Age 35-39', fontsize=20)

    

plot_country_single(largest_trend, 'Men', 'Single') 

plt.legend(largest_trend, loc='upper left', fontsize=14)
plt.suptitle('Largest Single Trend Nations Single Percentage for Men', fontsize=28)
plt.show()
```

Here we have the data looking at men. There is an obvious upward trend at some age range except United Arab Emirate. Of note, it appears Madives and United Arab Emirates it is just happening later in life but has stayed close to even levels 30+. 

```python
plot_country_single(largest_trend, 'Women', 'Single')

plt.legend(largest_trend, loc='upper left', fontsize=14)
plt.suptitle('Largest Single Trend Nations Single Percentage for Women', fontsize=28)
plt.show()
```

Here with Women's marriage percentage we can begin to see why United Arab Emirates is included. Not shown, but of note, is also a giant jump in the 15-19 age range of 43.5% in 1975 to 81.5% in 1987 for Women in the United Arab Emirates and from 53.1% in 1995 to 93.3% in 2006 for Women in Maldives.

Another interesting groupings, Hungry and Brazil both a massive upward trend in the 00's of many 30-39 year olds staying 'single'. Though Sweden either leadsor is close to percentage-wise for the highest percentage of single people in every graph.

```python
un_gni_lim= un_gni[un_gni['Year']<2010]
fig = plt.figure(figsize=(15,10))
for country in largest_trend:
    plt.plot(un_gni_lim[un_gni_lim['Country or Area'] == country]['Year'], un_gni_lim[un_gni_lim['Country or Area'] == country]['Value'])

plt.xlabel('Year')
plt.ylabel('GNI per Capita (current USD)')
plt.legend(largest_trend, loc='upper left', fontsize=14)
plt.suptitle('GNI of Countries with Largest trend of Singles', fontsize=28)
plt.show()
```

In this graph we can see two inital economic groupings that stay relatively closer to each other. The first (Maldives, Hungary, and Brazil) all begin below the 50 percentile in the 70's and by the end of the 00s they are close to the mean (~70 percentile). Later Czech Republic joins this group and has a similar trajectory.

The second group of countries (Sweden, Hungary, Belgium, United Arab Emirates, Kuwait, and Bahamas) all start out above the decades mean in the 1970's. These countries were already some of the more economically advanced countries and, much like their counter parts in the other group, have continued to outpace the gross average rate of growth. 

Now, let's take a look at some of the countries that have not seen the same type of increase in single rates amongst young people and see if there are any other connections we might be able to infer.  

```python
reverse_trend = ['Belize', 'Saint Lucia', 'Saint Vincent and the Grenadines', 'Mozambique', 'Ecuador', 'Guyana', 'Haiti']

plot_country_single(reverse_trend, 'Men', 'Single')

plt.legend(reverse_trend, loc='upper left', fontsize=14)
plt.suptitle('Reverse Single Trend Nations Single Percentage for Men', fontsize=28)
plt.show()
```

Here it is easy to see that these trends are not increasing like the previous group. Several of these countries (Belize, Saint Lucia, and Saint Vincent) all had high single rates in the 30-39 age range until the 90's. Part of their decline could be explained by regressing to the mean.

```python
plot_country_single(reverse_trend,'Women', 'Single')

plt.legend(reverse_trend, loc='upper left', fontsize=14)
plt.suptitle('Reverse Single Trend Nations Single Percentage for Women', fontsize=28)
plt.show()
```

Here we see much of the same as with the men in this grouping and we can note how similar the percentages are between men and women in the 30-39 age range compared to the large trend group.

```python
fig = plt.figure(figsize=(15,10))
for country in reverse_trend:
    plt.plot(un_gni_lim[un_gni_lim['Country or Area'] == country]['Year'], un_gni_lim[un_gni_lim['Country or Area'] == country]['Value'])

plt.xlabel('Year')
plt.ylabel('GNI per Capita (current USD)')
plt.legend(reverse_trend, loc='upper left', fontsize=14)
plt.suptitle('GNI of Countries with Reverse Trend of Singles', fontsize=28)
plt.show()
```

Here we get a better picture as to how GNI and marriage correlate. While all countries in large trend group have outpaced global economic growth, the best of these countries are still falling about 50% short of the mean at best (except late 70's early 80's Ecuador). Furthermore, all these economies are small.  Mozambique has the largest population ranking it ~ 50th largest population in the world. Haiti, Maldives, Saint Lucia, and Saint Vincent are all islands. The other interesting grouping is South American/Carribean countries. Only Mozambique and Maldives are not in that area of the world in this reverse trend group. 

It appears that the strongest correlation between marriage and GNI we can assert from the tools we have used is that if a country has outpaced global growth they are more likely to also be staying single longer. Obviously, this isn't an absolute as Maldives, Saint Lucia, and Saint Vincent have had similar GNI growth over the past 4 decades but have varying trends.
