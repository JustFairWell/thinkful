```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

wine_review = pd.read_csv('/Users/justinsmith/Documents/GitHub/wine-reviews/winemag-data-130k-v2.csv')
un_gni = pd.read_csv('/Users/justinsmith/Documents/GitHub/thinkful/un_gni.csv')
```

```python
brazil_gni = un_gni[un_gni['Country or Area'] == 'Brazil']
brazil_gni.plot(x='Year', y='Value')
plt.ylabel(s='GNI')
plt.show()

# Straight forward graph line plot. Good for overview but not very descriptive.
```

```python
plt.figure(figsize=(9,6))
sns.boxplot(x='Year', y='Value', data=brazil_gni)
plt.xticks(rotation=45)
plt.ylabel('GNI')
plt.show()

# Not really the best use of a box plot. Basically looks like a scatterplot lacking variables.
```

```python
plt.scatter(x='Year', y='Value', data=brazil_gni)
plt.show()

# Essentially the same as the last plot since box plots had only one variabe per year.
# To me, this is slightly better than line plot  since it is easier to see exact point
```

```python

sns.lmplot(x='Year', y='Value', data=brazil_gni)
plt.ylim(bottom=0)
plt.show()

#  Same as the scatterplot before but this time the a regression line.
# regression line would be better if put on a log scale since it goes negative.
```

```python
wine_89n90 = wine_review.loc[wine_review['points'].isin(range(89,91))]

sns.boxplot(x='points', y='price', data=wine_89n90)
plt.show()

# Here, the boxplot is much more useful because of the range of values. Clearly shows range with outliers.
```

```python
sns.barplot(x='points', y='price', data=wine_89n90)
plt.show()

# This barplot shows better scale of what wines typically at that score cost 
# (avg(90 points cost) > avg (89 point cost)) but it lacks the detail of the outliers.
```

```python
sns.violinplot(x='points', y='price', data=wine_89n90)
plt.show()

# The violion plot is a nice combination of the other two. You can see the slightly higher and tighter
# cluster of the 90 point wines, though the bar graph still shows this better. Here you get full range
# of values though like the box plot. 
```

```python
sns.barplot(x='country', y='price', data=wine_review)
plt.xticks(rotation=90)
plt.show()

#  Here you can see the average cost of wine divided by country. 
#  Again, you can't see outliers but good visual for average.
#  The varying colors helps make this easy to see. 
```

```python
sns.lvplot(x='country', y='price', data=wine_review)
plt.xticks(rotation=90)
plt.show()

# Here we have most of the perks of the bar plot above but with the ability to see outliers. 
# The average price looks flatter like this though the bar graph shos average price can double. 
```

```python
sns.pointplot(x='country',y='price', data=wine_review)
plt.xticks(rotation=90)
plt.show()

# This graph would work better for ranked categorical data i.e. based on wine points. 
# For these types of categories a point plot lie really isn't useful and suffers from sample size
# per country ie. Switzerland.
```

```python
sns.stripplot(x='country',y='price', data=wine_review)
plt.xticks(rotation=90)
plt.show()

# Very similar to the bar, and lv plots, this just gives more detail each value has point. 
# For the number of points in each category this is overkill and really doesn't vary much from lv other 
# than the estimate points
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="country", y="price", data=wine_review)
plt.xticks(rotation=90)
plt.show()

# Looks very similar to bar, strip, and lv plot. As a pracitcal matter, the boxes here because of 
# the range, doesn't give you any information. This would work better are short ranges of values like
# wine_89n90.
```

```python
plt.figure(figsize=(20, 10))
sns.violinplot(x="country", y="price", data=wine_review, hue='country')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xticks(rotation=90)
plt.show()

# This suffers the same fate as the boxplot. The violin shapes are too small to be useful 
# even after getting the figure to a much bigger size. and same as with box plot, 
# it would be better with closer range of values.
```
