```python
import numpy as np
import matplotlib.pyplot as plt

normal_distrib = np.random.normal(5, 0.5, 100)
normal_distrib2 = np.random.normal(10, 1, 100)
norm_distrib_combo = normal_distrib + normal_distrib2
norm_mean = np.mean(norm_distrib_combo)
norm_std = np.std(norm_distrib_combo)
plt.hist(norm_distrib_combo)
plt.axvline(norm_mean, color='b', linestyle='dashed')
plt.axvline(norm_mean - norm_std, color='r', linestyle='dotted')
plt.axvline(norm_mean + norm_std, color='r', linestyle='dotted')
plt.show()
```

```python
# This is the combination graph of the two normal distributions. Here, the mean and std are very useful.
# Most point are within one std and almost all are within two std. 
```

```python
poisson_distrib = np.random.poisson(10, 100)
poisson_mean = np.mean(poisson_distrib)
poisson_std = np.std(poisson_distrib)
plt.hist(poisson_distrib)
plt.axvline(poisson_mean, color='b', linestyle='dashed')
plt.axvline(poisson_mean - poisson_std, color='r', linestyle='dotted')
plt.axvline(poisson_mean + poisson_std, color='r', linestyle='dotted')
plt.show()
```

```python
# This is a poisson distribution. The mean and std do not provide much helpful information.
# Most of the points in this graph do fall within two std and most are close to th mean.
```

```python
lognorm_distrib = np.random.lognormal(10, 1, 100)
lognorm_mean = np.mean(lognorm_distrib)
lognorm_std = np.std(lognorm_distrib)
plt.hist(lognorm_distrib)
plt.axvline(lognorm_mean, color='b', linestyle='dashed')
plt.axvline(lognorm_mean - lognorm_std, color='r', linestyle='dotted')
plt.axvline(lognorm_mean + lognorm_std, color='r', linestyle='dotted')
plt.show()
```

```python
# This is a normal log function. The mean and std are not very useful here. The mean is close to the origin
# because the groups are closer on that side of the mean with a steady decline as you increase the x axis. 
```

```python
gamma_distrib = np.random.gamma(10, 1, 100)
gamma_mean = np.mean(gamma_distrib)
gamma_std = np.std(gamma_distrib)
plt.hist(gamma_distrib)
plt.axvline(gamma_mean, color='b', linestyle='dashed')
plt.axvline(gamma_mean - gamma_std, color='r', linestyle='dotted')
plt.axvline(gamma_mean + gamma_std, color='r', linestyle='dotted')
plt.show()
```

```python
laplace_distrib = np.random.laplace(10, 1, 100)
laplace_mean = np.mean(laplace_distrib)
laplace_std = np.std(laplace_distrib)
plt.hist(laplace_distrib)
plt.axvline(laplace_mean, color='b', linestyle='dashed')
plt.axvline(laplace_mean - laplace_std, color='r', linestyle='dotted')
plt.axvline(laplace_mean + laplace_std, color='r', linestyle='dotted')
plt.show()
```

```python
# This is a laplace distribution. Much like the previous distribution, the mean and std aren't extremely useful.
# Here they are somewhat more predictive since the division of the points is more evenly divided on both
# sides of the mean. Still some serious outlyiers with a power distribution.
```

```python
hyper_distrib = np.random.hypergeometric(10, 10, 10, 100)
hyper_mean = np.mean(hyper_distrib)
hyper_std = np.std(hyper_distrib)
plt.hist(hyper_distrib)
plt.axvline(hyper_mean, color='b', linestyle='dashed')
plt.axvline(hyper_mean - hyper_std, color='r', linestyle='dotted')
plt.axvline(hyper_mean + hyper_std, color='r', linestyle='dotted')
plt.show()
```

```python
# This is a hypergeometric distribution. Here the mean and std are very telling of the data.
# Since this graph is meant to be more geometrical, the mean and std almost perfectly discribe the data. 
```

```python
pareto_distrib = np.random.pareto(10, 100)
pareto_mean = np.mean(pareto_distrib)
pareto_std = np.std(pareto_distrib)
plt.hist(pareto_distrib)
plt.axvline(pareto_mean, color='b', linestyle='dashed')
plt.axvline(pareto_mean - pareto_std, color='r', linestyle='dotted')
plt.axvline(pareto_mean + pareto_std, color='r', linestyle='dotted')
plt.show()
```

```python
# Ths is a pareto distribution. Much like the normal log distribution, the mean and std are not very telling
# because the information is lopsided, with a high concentration of points closer to the origin.
```
